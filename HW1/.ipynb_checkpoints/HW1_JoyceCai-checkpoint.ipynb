## Task4: make Matplolib figures and save

```python
# set default parameters
params = {'legend.fontsize': 14, \
          'xtick.labelsize':14, \
          'ytick.labelsize':14, \
          'font.size':14}
plt.rcParams.update(params)
```

```python
# generate random fields
from numpy import random

N = 100
noise = random.rand(N)-0.5 # Gaussian distribution
```

```python
x = np.arange(N)
y = 2*np.sin(x)+noise

plt.plot(x,y)
plt.xlabel('Timestep')
plt.ylabel('Amplitude')
plt.grid(True)
plt.title('Sine Wave Signal w/ Random Noise')
plt.show()
```

```python
plt.savefig('plot_wave.png',dpi=300) # add resolution with dpi
```

```python
# Fast Fourier Transform to look at the freatures of the wave
from scipy.fft import rfft, rfftfreq

yf = rfft(y)
xf = rfftfreq(N, 1)

plt.plot(xf, np.abs(yf))
plt.xlabel('Frequency')
plt.ylabel('Spectrum')
plt.title('Frequency Spectrum From FFT')
plt.grid(True)
plt.show()
```

```python
plt.savefig('plot_wave_fft.png',dpi=300) # add resolution with dpi
```
